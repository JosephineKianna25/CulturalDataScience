## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

### Activity 1 + 2

```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")

# Jeg laver to dataframes der skal indeholde data fra hhv. fysiske og digitale best selling seingler
df_physical = rawdata[0]
df_digital = rawdata[3]

# Jeg tilføjer nye kolonner der angiver om en single er fysisk eller digital
df_physical['Type'] = 'Physical Single' 
df_digital['Type'] = 'Digital Single'

# Til sidst kombinerer jeg det hele, og bruger pd.concat() for at kombinere dem, 
# samt axis=1 for at tilføje dem side om side.
best_singles = pd.concat([df_physical, df_digital], axis=0)

selected_columns = ['Single', 'Artist', 'Released', 'Type'] # bare for det visuelle
best_singles_cleaned = best_singles[selected_columns]

print(best_singles_cleaned)
```

### Activity 3: Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.

```python
birthyear = 1999

best_singles.loc[:, 'Year Difference'] = (best_singles['Released'] - birthyear).abs()

# best_singles_cleaned - birthyear beregner forskellen mellem alle udgivelsesår og birthyear
# resultat bliver en Serie med forskelle.abs() anvendes til at tage den absolutte værdi af 
# forskellene, så der kommer positive tal, uanset om sangen er udgivet før eller efter 1999

# værdierne tildeles den nye kolonne 'Year Difference' fra den beregnede absolutte forskel 
# dette gør vi med .loc,  der tildeler værdierne. .loc er også en metode der bruges til at
# indeksere dataframes. Det giver mulighed for at tilgå rækker og kolonner vha labels. 
# Her bruges den til at specificere at vi ønsker at opdatere kolonnen 'year difference'
# for alle rækker - : betyder alle rækker.

closest_year = best_singles['Year Difference'].min()
# finder det nærmeste år, dvs. det med den mindste værdi i kolonnen 'Year Difference'

closest_singles = best_singles.loc[best_singles['Year Difference'] == closest_year]
# bruges til at finde alle sange hvor 'year difference' er lig med den tætteste værdi
# giver en dataframe med sange der blev udgivet i det år, der er tættest på fødselsåret 

df_closest_singles = pd.DataFrame(closest_singles)

df_closest_singles
```

### Activity 4: Outputs the artist and single name for the year you were 15 years old.

```python
songs_2014 = best_singles[best_singles['Released'] == 2015]
# jeg tilgår release-rækken og tjekker efter tilfælde
songs_2014
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


### Activity 1: Make a single dataframe that combines the space missions from the 1950's to the 2020's

```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")

df_space_missions = pd.concat(rawdata[0:8]) # 0-8 marks 1950's to 2020's

df_space_missions
```

### Activity 2: Write a script that returns the year with the most launches

```python
# Jeg laver et if-statement til at tjekke kolonnen 'Launch date' i df'en (og om den findes) 
if 'Launch date' in df_space_missions.columns: #df_space_missions.columns returnerer en liste over kolonnenavne i df
    launch_counts = df_space_missions['Launch date'].value_counts()
    # value_counts() er en pandas-metode - den tæller antallet af forekomster af hver unik værdi i 'launch date' 
    # resultatet er en Series, hvor indekset er de forskellige datoer - værdien er antallet af gange, hver dato
    # optræder i df. Series gemmes i variablen launch_counts. 
    most_launches_year = launch_counts.idxmax() #idmax() finder indekset (datoen) for max værdi i launch_counts
    max_launches = launch_counts.max() # max() returnerer max værdi fra launch_counts, der er det højeste antal
    # launches for en enkelt dato.

print(f"The year with the most launches is {most_launches_year} with {max_launches} launches.")
    
```

### Activity 3: Write a script that returns the most common month for launches

```python
if 'Launch date' in df_space_missions.columns: #  tjekker efter en dato kolonnne (der hedder launch date)
        df_space_missions['Launch date'] = pd.to_datetime(df_space_missions['Launch date'])
        # konverterer launch date til datetime format 

        df_space_missions['Month Name'] = df_space_missions['Launch date'].dt.month_name()
        # tager måneden fra 'launch date'
        month_counts = df_space_missions['Month Name'].value_counts()
        # .dt.month_name() får den til at skrive navn på måned, istedet for fx 11 for november

        most_common_month = month_counts.idxmax()
        max_launches = month_counts.max()
        # samme funktioner som i aktivitet 2

print(f"The most common month for launches is {most_common_month} with {max_launches} launches.")
```

### Activity 4: Write a script that ranks the months from most launches to fewest launches

```python
# jeg bruger .sort_values til at sortere min dataframe 

df_months_sorted = month_counts.sort_values(ascending = False) 
# ascending - true eller false afgør om sortering skal være stigende eller faldende. 
# true = stigende (mindst til størst). false = faldende (størst til mindst).
df_months_sorted
```

## Supervillain challenge

1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

### Activity 1: Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")

df_supervillains = pd.concat(rawdata[1:12])
df_supervillains
```

### Activity 2: Write a script that ranks each decade in terms of how many supervillains debuted in that decade

```python
df_supervillains['Year Debuted'] = pd.to_numeric(df_supervillains['Year Debuted'], errors='coerce')
# Først sikrer jeg mig at kolonnen 'Year Debuted' kun indeholder numeriske værdier vha. pd.to_numeric()
# errors = 'coerce' betyder at hvis der findes ugyldige værdier i kolonnen, bliver disse 
# erstattet med NaN - hjælper med at undgå fejl senere i koden 

df_supervillains_clean = df_supervillains.dropna(subset=['Year Debuted']).copy()
# her fjernes alle rækker i 'Year Debuted' med værdien NaN, så der kun
# arbejdes med gyldige tal 
# Jeg laver en kopi af df'en, for at undgå SettingWithCopyWarning

df_supervillains_clean['Decade'] = (df_supervillains_clean['Year Debuted'] // 10) * 10
# Her omdannes år til årtier, første ved at dividere med // (heltalsdivision), der 
# fjerner decimalerne. F.eks. 1973 // 10 = 197. Derefter ganger vi med 10 og får 
# 1970 - og så kan vi se at 1973 tilhører 1970'erne. Det samme gælder de andre årstal. 
# De firkantede parenteser i variabler opretter desuden den nye variabel 'Decade'

decade_counts = df_supervillains_clean.groupby('Decade').size()
# groupby('Decade') grupperer df'en efter den nye kolonne med årtier
# size() tæller antal af række i hver gruppe - antallet af supervillain debuter i hvert årti
# giver en Series hvor indekset er årtiet - værdien er antal af debuter i det respektive årti

decade_ranking = decade_counts.sort_values(ascending=False)
# Sorterer tællingerne i forrige trin i faldende rækkefølge

decade_ranking = pd.DataFrame(decade_ranking)
# laver det til en dataframe for at gøre det pænt

decade_ranking
```

### Activity 3: Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
if 'Company' in df_supervillains_clean.columns:
    # Fjern dubletter for at sikre unikke supervillains
    unique_supervillains = df_supervillains_clean.drop_duplicates(subset=['Company', 'Character / Team'])

    # Value counts tiæller forekomster af hvert company i den filterede df unique_supervillains
    company_counts = unique_supervillains['Company'].value_counts()

    company_df = company_counts.reset_index()
    # Jeg omdanner company_counts til en DataFrame og resetter indeks
    # Første kolonne bliver indekset
    
    company_df.columns = ['Company', 'Number of Supervillains']
    # Giver kolonnerne navne

    company_df = company_df.sort_values(by='Number of Supervillains', ascending=False).reset_index(drop=True)
    # Sorterer værdierne i faldende rækkefølge
    
company_df

```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df.head()

# Jeg laver først en dictionary der bruges til at omdøbe de forskellige opgaver
# fra deres tekniske navne til forståelige beskrivelser
task_mapping = {
    'Reading_NoInt': 'Reading with No Interference',
    'Naming_Int': 'Naming with Interference',
    'Naming_NoInt': 'Naming with No Interference',
    'Reading_Int': 'Reading with Interference'
}

# Jeg bruger pd.melt() til at ændre DF fra bred til lang format
df_long = df.melt(
    id_vars=['Year'],  # Jeg angiver at kolonnen Year skal bevares i den nye df
    value_vars=list(task_mapping.keys()),  # Her angiver jeg hvilke kolonner
    # jeg ønsker at smelte sammen 
    var_name='Task',  # Den nye kolonne til opgaverne der lige er oprettet får navnet Task 
    value_name='RT'   # Den nye kolonne for reaktionstid skal hedde RT
)

df_long['Task'] = df_long['Task'].map(task_mapping)
# Opgaverne omdøbes vha. ordborgen task_mapping

df_long['Participant_id'] = df_long.index + 1
# Her tilføjes deltager ID (+1 sørger for at den starter fra 1)

df_long = df_long[['Participant_id', 'Year', 'Task', 'RT']]
# Kolonnerne omorganiseres

print(df_long.head())

# Kan ikke finde ud af at fjerne 'default' indekseringen - spørg om hjælp
# Evt genlæs kap. 7

################################################################

# Kan også gøres sådan her, sofies kode: 

df['Participant_id'] = df.index +1 

df_long_2 = pd.melt(df,
                  id_vars=['Participant_id', 'Year'],
                  var_name='Task',
                  value_name='RT')

print(df_long_2.head())

# var_ er en specifik funktion i melt() der fortæller hvilke kolonner der skal bevares
# og som skal indeholde de smeltede kolonner ? 
```
